```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
import library
import matplotlib.patches as mpatches
```

```python
data = library.loadList('exoplanet.csv')
col = {'mass':2,'rad':3,'period':4,'sma':5}
```

```python
firstMaster = []
for n in range(len(data)):
    radius = data[n][col['rad']]
    sma = data[n][col['sma']]
    mass = data[n][col['mass']]
    if radius != '' and sma != '':
        radius = float(radius)
        sma = float(sma)
        if mass == '':
            mass = 0
        mass = float(mass)
        firstMaster += [[round(radius,2),round(sma,2),round(mass,2)]]

firstMaster.sort()
print(firstMaster)
```

```python
print('Masses are automatically zero in the case of no-data:')
df1 = pd.DataFrame(firstMaster,columns=['radius','sma','mass'])
df1
```

```python
secondMaster = []
for n in range(len(data)):
    radius = data[n][col['rad']]
    period = data[n][col['period']]
    mass = data[n][col['mass']]
    if radius != '' and period != '':
        radius = float(radius)
        period = float(period)
        if mass == '':
            mass = 0
        mass = float(mass)
        secondMaster += [[round(radius,2),round(period,2),round(mass,2)]]

secondMaster.sort()
print(secondMaster)
```

```python
print('Masses are displayed as zero for unknown values:')
df2 = pd.DataFrame(secondMaster,columns=['Radius','Period','Mass'])
df2
```

```python
#Function for colour
def getColor(mass):
    cList = []
    for n in range(len(mass)):
        m = mass[n]
        if m == 0:
            color = 'blue'
        else:
            if m < 0.5:
                color = 'green'
            elif m >= 0.5 and m < 1:
                color = 'yellow'
            else:
                color = 'red'
        cList.append(color)
    return(cList)

#Sizes
plt.figure(figsize=(15,5))

plt.subplot(121)
x = df1.sma
y = df1.radius
mass = df1.mass
c = getColor(mass)
scatter = plt.scatter(x,y,c=c,s=15)
plt.title('Radius vs. Semi-Major-Axis',fontweight='bold',fontsize=15)
plt.xlabel('Semi-Major-Axis (AU)',fontweight='bold')
plt.xlim(0,1)
plt.ylabel('Radius (Jpt.)',fontweight='bold')
plt.ylim(0,2.5)
blue_patch = mpatches.Patch(color='blue', label='unknown')
green_patch = mpatches.Patch(color='green', label='low')
yellow_patch = mpatches.Patch(color='yellow', label='medium')
red_patch = mpatches.Patch(color='red', label='high')
legend = plt.legend(title='Mass',handles=[blue_patch,green_patch,yellow_patch,red_patch],loc=1)

plt.subplot(122)
x = df2.Period
y = df2.Radius
mass = df2.Mass
c = getColor(mass)
plt.scatter(x,y,c=c,s=15)
plt.title('Radius vs. Period',fontweight='bold',fontsize=15)
plt.xlabel('Period (Days)',fontweight='bold')
plt.xlim(0,100)
plt.ylabel('Radius (Jpt.)',fontweight='bold')
plt.ylim(0,2)
blue_patch = mpatches.Patch(color='blue', label='unknown')
green_patch = mpatches.Patch(color='green', label='low')
yellow_patch = mpatches.Patch(color='yellow', label='medium')
red_patch = mpatches.Patch(color='red', label='high')
legend = plt.legend(title='Mass',handles=[blue_patch,green_patch,yellow_patch,red_patch],loc=1)

plt.suptitle('Grouping Scatter Plots (Looking for General Groups of typical Exoplanets)',fontweight='bold',
             fontsize=20,y=1.05)
plt.show()
```
